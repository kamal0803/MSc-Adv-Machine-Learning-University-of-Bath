```python
x = np.arange(1, 20.1, 0.1)
```

```python
y = []
```

```python
for i in x:
    y.append(3*i + 2 + np.random.randn())
```

```python
y = np.array(y)
```

```python
X = np.c_[np.ones(len(x)), x] 
y = y.reshape(-1,1)
```

### Batch gradient descent

```python
terminal = False
alpha = 0.01
i = 0
w_t = np.random.randn(2).reshape(-1,1)

print(w_t.shape)
```

```python
n = X.shape[0]
while not terminal:

    y_pred = X @ w_t
    loss = (1 / (2 * n)) * np.sum((y - y_pred)**2)
    
    gradient = (1/n) * X.T @ (X @ w_t - y)
    w_t = w_t - (alpha)*gradient

    #print(f"Loss: {loss}")
    
    if np.linalg.norm(gradient) < 1e-06:
        terminal = True
```

### Stochastic Gradient Descent

```python
terminal = False
alpha = 0.01
i = 0
w_t = np.random.randn(2).reshape(-1,1)

while not terminal:

    row = np.random.randint(0, X.shape[0])

    loss = 0.5 * (y[row] - y_pred[row])**2
    gradient = X[row].T.reshape(2,1) @ (X[row].reshape(1, 2) @ w_t.reshape(-1, 1) - y[row].reshape(1, 1))
    w_t = w_t - alpha*gradient

    #print(f"Loss: {loss}")
    
    i = i + 1
    alpha = alpha * 0.999
    
    if i > 100000:
        terminal = True
```

### Mini Batch Gradient Descent

```python
def generate_batches(mini_batch_size=32):

    j = 0
    mini_batches = []
    
    while j < mini_batch_size:
        row = np.random.randint(0, X.shape[0])
        if row not in mini_batches:
            mini_batches.append(row)
            j = j + 1
    
    mini_batches.sort()

    h = []
    for i in mini_batches:
        h.append(X[i])

    X_mbgd = np.vstack(tuple(h))

    h = []
    for i in mini_batches:
        h.append(y[i])
    
    y_mbgd = np.vstack(tuple(h))

    return X_mbgd, y_mbgd
```

```python
terminal = False
alpha = 0.01
i = 0
w_t = np.random.randn(2).reshape(-1,1)

while not terminal:


    X_mbgd, y_mbgd = generate_batches()
    n = X_mbgd.shape[0]

    loss = (1 / (2 * n)) * np.sum((X_mbgd @ w_t - y_mbgd)**2)

    #print(f"Loss: {loss}")
    
    gradient = (1/n) * X_mbgd.T @ (X_mbgd @ w_t - y_mbgd)
    w_t = w_t - (alpha)*gradient
    i = i + 1
    
    if np.linalg.norm(gradient) < 1e-06 or i > 100000:
        terminal = True
```
