```python
import numpy as np
```

```python
def gradient(x_t):
    
    return np.array([2*x_t[0], 2*x_t[1]])
```

```python
alpha = 0.1
```

```python
terminal = False
i = 0
x_t = np.array([2,2])

while not terminal:

    derivative = gradient(x_t)
    x_t_plus_one = x_t - alpha*derivative
    x_t = x_t_plus_one
    print(x_t)
    i = i + 1

    if i == 1000 or np.linalg.norm(derivative) < 1e-06:
        terminal = True
```
