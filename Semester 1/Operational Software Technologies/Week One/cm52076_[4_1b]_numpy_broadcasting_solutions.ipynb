## Exercise 1 — Scalar Broadcasting (Solution)

```python

import numpy as np

# Define array and scalar
x = np.array([1,2,3])
y = 5
z = x + y

print("x:", x)
print("y:", y)
print("Result:", z)

```

## Exercise 2 — Matrix + Vector Broadcasting (Solution)

```python

a = np.array([[1,2,3],
              [4,5,6]])
b = np.array([10,20,30])

c = a + b
print("a:\n", a)
print("b:\n", b)
print("Result:\n", c)
```

## Exercise 3 — Row + Column Broadcasting (Solution)

```python

col = np.array([[1],[2],[3]])
row = np.array([10,20,30,40])

result = col + row
print("col:\n", col)
print("row:\n", row)
print("Result:\n", result)

```

## Exercise 4 — Broadcasting Error (Solution)

```python

a = np.ones((3,2))
b = np.ones((2,3))

print("a:\n", a)
print("b:\n", b)
# This will raise an error due to incompatible shapes
try:
    res = a + b
except ValueError as e:
    print("Broadcasting Error:", e)

#Solution - match the dimensions of either a or b by taking transpose.

a = np.ones((3,2)).T
print("a:\n", a)
print("b:\n", b)
res = a + b
print("Result:\n", res)
```
