## Exercise 1 — Scalar Broadcasting
Add a scalar to a NumPy array.

```python
import numpy as np
```

```python
x = np.array([1,2,3])
y = 4   # choose a scalar
z = x + y   # add scalar to array

print(z)

```

## Exercise 2 — Matrix + Vector Broadcasting
Add a row vector to a 2D matrix.

```python

A = np.array([[1,2,3],
              [4,5,6]])
b = np.array([10,20,30])

c = A + b   # broadcast row vector

print(c)
```

## Exercise 3 — Row + Column Broadcasting
Perform an outer sum using a column and row vector.

```python

col = np.array([[1],[2],[3]])
row = np.array([10,20,30,40])

result = col + row

print(result)
```
