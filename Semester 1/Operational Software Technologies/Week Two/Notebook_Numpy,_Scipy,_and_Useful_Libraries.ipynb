## 1 NumPy Overview

**NumPy** (Numerical Python) is the **fundamental package for scientific computing in Python**. It provides support for **multidimensional arrays (ndarray)**, along with a large collection of mathematical functions to operate on them efficiently.  

Unlike Python lists, NumPy arrays are **fast, memory-efficient, and vectorized**, making them the backbone of numerical analysis, data science, and machine learning workflows.  

Researchers and developers use NumPy for:  
- **Linear algebra** (matrix multiplication, eigenvalues).  
- **Statistical operations** (mean, median, standard deviation).  
- **Array broadcasting** (operations on arrays of different shapes).  
- **Fast elementwise math** (sin, log, exp, etc.).  
- **Random number generation**.  

---

### Commonly Used Functions and Features

- **Array creation**: `np.array`, `np.zeros`, `np.ones`, `np.arange`, `np.linspace`  
- **Statistics**: `np.mean`, `np.std`, `np.median`, `np.sum`  
- **Linear algebra**: `np.dot`, `np.matmul`, `np.linalg.inv`  
- **Random**: `np.random.rand`, `np.random.randn`, `np.random.randint`  
- **Broadcasting**: arithmetic on arrays of different shapes

### 1.1 Creating and manipulating data structures

```python
# Array creation
a = np.array([1, 2, 3, 4])
print("\nArray a =", a)

b = np.arange(1, 13, 2)   # 6 elements: [1,3,5,7,9,11]
print("\nArray b (arange) =", b)

# Reshape into a 3×2 matrix
M = b.reshape(3, 2)
print("\nMatrix M =", M)

# Statistics
print("\nMean of M =", np.mean(M))
print("\nStandard deviation of M =\n", np.std(M))

# Linear algebra
x = np.array([1, 2])
y = np.array([3, 4])
print("\nDot product =", np.dot(x, y))

# Broadcasting example
vec = np.array([1, 2, 3])
mat = np.ones((3, 3))
print("\nBroadcasting mat + vec =\n", mat + vec)

# Random numbers
print("\nRandom uniform [0,1) =", np.random.rand(3))
print("\nRandom normal samples =", np.random.randn(3))
print("\nRandom integers 0–9 =", np.random.randint(0, 10, size=5))
```

### 1.2 Sorting using Numpy

### 1.3 Filtering using Numpy
Let's explore filtering (masking) in Numpy

```python
#Check if any element meets a condition
print("\n(a > 0.1).any() = ",(a > 0.1).any())

#Check if all element meets a condition
print("\n(a > 0.1).all() = ",(a > 0.1).all())

#Filter (masking) elements that meet a condition
mask = a > 0.1
print("\na > 0.1 = ",mask)
print("\na[(a > 0.1)] = ",a[mask])

#Filter (masking) elements that meet multiple conditions
mask = np.logical_and(a>0.2, a<0.8)
print("\n(a>0.2) & (a<0.8) = ",mask)
print("\na[(a>0.2) & (a<0.8)] = ",a[mask])
```

### 1.4 Random Number Generation with `numpy.random`

This example demonstrates how to use **`numpy.random`** to generate random numbers, shuffle arrays, draw samples from distributions, and control reproducibility with **random seeds**.




```python

# Generate a length-8 vector of random numbers
# drawn from the uniform distribution on [0, 1).
a = np.random.rand(8)
print("\na = ", a)

# Shuffle the array in place (random order).
np.random.shuffle(a)
print("\nnp.random.shuffle(a) = ", a)

# Generate an 8×8 matrix of samples from the standard normal distribution
# (mean = 0, standard deviation = 1).
normal_samples = np.random.standard_normal(size=(8,8))
print("\nnp.random.standard_normal(size=(8,8)) = ", normal_samples)

# Draw 6 samples from a Beta distribution with α=2.0 and β=4.0.
beta_samples = np.random.beta(2.0, 4.0, size=6)
print("\nnp.random.beta(2.0, 4.0, size=6) = ", beta_samples)

# --- Random seed for reproducibility ---

# Reset the seed to 12 and draw 8 random numbers.
np.random.seed(12)
print("\nRandom numbers with seed 12 (first draw) = ", np.random.rand(8))

# Reset the seed again to 12 → produces the same sequence.
np.random.seed(12)
print("\nRandom numbers with seed 12 (second draw) = ", np.random.rand(8))

# Changing the seed to 13 produces a different sequence.
np.random.seed(13)
print("\nRandom numbers with seed 13 = ", np.random.rand(8))
```

### 1.5 Derivatives and Integrals (Numerical Methods)

We can approximate **derivatives** and **integrals** using **NumPy** arrays. These approximations are based on simple but powerful ideas from calculus:

- A **[Derivative – Wikipedia](https://en.wikipedia.org/wiki/Derivative)** estimates the rate of change or slope of a function at a point.  

- An **[Integral – Wikipedia](https://en.wikipedia.org/wiki/Integral)** estimates the area under a curve between two points.  

#### Derivatives: Finite Difference Approximation

To approximate the derivative $f'(x)$, we use the **central difference**:

$
f'(x) \approx \frac{f(x+h) - f(x-h)}{2h}
$

This works well on a grid of values (arrays) and is easy to implement using NumPy.

#### Integrals: Trapezoidal Rule

To approximate the area under a curve, we use the **trapezoidal rule**, which approximates the area as a sum of trapezoids:

$
\int_a^b f(x) dx \approx \sum \tfrac{1}{2}(y_i + y_{i+1})(x_{i+1} - x_i)
$

NumPy provides a built-in function: `np.trapezoid(y, x)`  


```python
# Define a smooth function: sin(x)
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)

# Estimate derivative using central differences
h = x[1] - x[0]
dy_dx = (np.roll(y, -1) - np.roll(y, 1)) / (2 * h)

# Estimate integral using trapezoidal rule
integral = np.trapezoid(y, x)

# Plot original function and its derivative
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.plot(x, dy_dx, label='Approx. derivative', linestyle='--')
plt.title('Derivative using Central Differences')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.fill_between(x, y, alpha=0.3)
plt.title(f'Integral using Trapezoidal Rule ≈ {integral:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### 1.6 Linear Algebra with `numpy.linalg`

The **`numpy.linalg`** module provides efficient tools for linear algebra.  
Here are some common operations with links for further reading:

- [Determinant](https://en.wikipedia.org/wiki/Determinant) → tells if a matrix is invertible and measures volume scaling.  
- [Matrix inverse](https://en.wikipedia.org/wiki/Invertible_matrix) → finds the inverse of a square matrix.  
- [Solving linear systems](https://en.wikipedia.org/wiki/System_of_linear_equations) → solve `Ax = b`.  
- [Eigenvalues and eigenvectors](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors) → important in stability analysis, PCA, physics.  
- [Singular Value Decomposition (SVD)](https://en.wikipedia.org/wiki/Singular_value_decomposition) → used in dimensionality reduction and data compression.  


```python
# Create a random 6×6 matrix 'a' and a 6×1 vector 'b'
a = np.random.rand(6, 6)
b = np.random.rand(6)

print("\na = ", a)
print("\nb = ", b)

# Determinant of matrix a
det_a = np.linalg.det(a)
print("\nDeterminant of a = ", det_a)

# Inverse of matrix a
inv_a = np.linalg.inv(a)
print("\nInverse of a = \n", inv_a)

# Solve linear system Ax = b
x = np.linalg.solve(a, b)
print("\nSolution to Ax = b = ", x)

# Eigenvalues and eigenvectors (for symmetric matrix a + a.T)
eigvals, eigvecs = np.linalg.eigh(a + a.T)
print("\nEigenvalues of (a + a.T) = ", eigvals)
print("\nEigenvectors of (a + a.T) = \n", eigvecs)

# Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(a)
print("\nSingular values of a = ", S)
```

### 1.7 Toy Example: Predicting Taxi Fare with `numpy.linalg`

Imagine you want to predict the **taxi fare (£)** based on one feature: the **distance travelled (km)**.  
We assume a linear model:

$
\text{fare} \approx \beta_0 + \beta_1 \cdot \text{distance}
$

We’ll fit this model using `numpy.linalg.lstsq` with a tiny dataset and then plot the fit.

```python
# Small dataset: distance (km) vs fare (£)
distance = np.array([1, 2, 3, 4])        # km
fare     = np.array([5, 6, 9, 12])       # £

# Build design matrix
X = np.c_[np.ones_like(distance), distance]
y = fare

# Solve least squares
beta, *_ = np.linalg.lstsq(X, y, rcond=None)
b0, b1 = beta

print(f"Estimated model: fare ≈ {b0:.2f} + {b1:.2f} × distance (km)")

# Predictions for plotting
x_line = np.linspace(0, 5, 100)
y_line = b0 + b1 * x_line

# Plot data and fitted line
plt.figure(figsize=(6,4))
plt.scatter(distance, fare, color="blue", label="Observed fares")
plt.plot(x_line, y_line, color="red", label="Fitted line")
plt.xlabel("Distance (km)")
plt.ylabel("Fare (£)")
plt.title("Taxi Fare vs Distance (Least Squares Fit)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Explanation
- **β₀ (intercept):** the **base fee** (charge just for starting the ride).  
- **β₁ (slope):** the **cost per kilometre**.  
- In this toy example the fit gives something close to:
  
  $
  \text{fare} ≈ 3 + 2 \cdot \text{distance}
  $
so the base fee is ~£3 and each km costs ~£2.  

This simple example demonstrates how `numpy.linalg.lstsq` can capture real-world linear cost structures.

### 2.2 `scipy.optimize`

This example shows how to use **[`scipy.optimize.minimize`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html)** to fit a straight line to data by minimizing a **cost function**.

- We have simple data points (x,y).  
- The model is a straight line:  
  $
  \hat{y} = \beta_0 + \beta_1 x
  $
- The **cost function** is the **[sum of squared errors](https://en.wikipedia.org/wiki/Least_squares)** between the predictions and the true values.  
- `scipy.optimize.minimize` adjusts the parameters $(\beta_0, \beta_1)$ to minimize the cost.

This is essentially **[linear regression](https://en.wikipedia.org/wiki/Linear_regression)**, but instead of using the closed-form formula, we directly solve it by **[numerical optimization](https://en.wikipedia.org/wiki/Mathematical_optimization)**.

```python
import scipy.optimize

# Example data points
x = np.array([0.0, 1.0, 2.0, 3.0])
y = np.array([1.0, 1.5, 3.2, 4.1])

# Define the cost function = sum of squared errors
def cost(param):
    intercept, slope = param
    estimate = intercept + slope * x         # predicted y values
    return np.square(estimate - y).sum()     # sum of squared errors

# Initial guess for [intercept, slope]
guess = np.array([0.0, 0.0])

# Run the optimization
res = scipy.optimize.minimize(cost, guess)

# Extract results
intercept, slope = res.x
print("\nFitted parameters [intercept, slope] = ", res.x)
print("\nFinal cost (sum of squared errors) = ", res.fun)

# Plot the data and fitted line
plt.figure(figsize=(6,4))
plt.scatter(x, y, color="blue", label="Data points")
plt.plot(x, intercept + slope * x, color="red", label="Fitted line")
plt.title("Line Fitting with scipy.optimize.minimize")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from scipy import stats

rng = np.random.default_rng(0)

# -----------------------------
# Normal distribution (PDF & CDF)
# -----------------------------
x_vals = np.linspace(-3, 3, 200)
pdf = stats.norm.pdf(x_vals, 0, 1)
cdf = stats.norm.cdf(x_vals, 0, 1)

# -----------------------------
# Binomial distribution (PMF)
# -----------------------------
n, p = 20, 0.3
k_vals = np.arange(0, n+1)
pmf = stats.binom.pmf(k_vals, n, p)

# -----------------------------
# t-test example groups
# -----------------------------
grp1 = rng.normal(0, 1, size=50)
grp2 = rng.normal(0.4, 1, size=55)

# -----------------------------
# Correlation example
# -----------------------------
x_corr = rng.normal(size=100)
y_corr = 0.7*x_corr + rng.normal(scale=0.7, size=100)

```

```python
# -----------------------------
# Plotting
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(12, 9))

# Normal distribution PDF & CDF
axes[0,0].plot(x_vals, pdf, label="PDF")
axes[0,0].plot(x_vals, cdf, label="CDF")
axes[0,0].set_title("Normal Distribution (μ=0, σ=1)")
axes[0,0].legend()
axes[0,0].grid(alpha=0.3)

# Binomial distribution PMF
axes[0,1].bar(k_vals, pmf, color="lightgreen", edgecolor="black")
axes[0,1].set_title(f"Binomial Distribution (n={n}, p={p})")
axes[0,1].set_xlabel("k")
axes[0,1].set_ylabel("PMF")
axes[0,1].grid(alpha=0.3)

# t-test group histograms
axes[1,0].hist(grp1, bins=15, alpha=0.6, label="Group 1")
axes[1,0].hist(grp2, bins=15, alpha=0.6, label="Group 2")
axes[1,0].set_title("Two Groups (for t-test)")
axes[1,0].legend()
axes[1,0].grid(alpha=0.3)

# Correlation scatter
axes[1,1].scatter(x_corr, y_corr, alpha=0.6)
axes[1,1].set_title("Correlation Example (Pearson r)")
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")
axes[1,1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### 4 `h5py` Overview

[`h5py`](https://www.h5py.org/) is a Python interface to the **HDF5** (Hierarchical Data Format v5) library.  
HDF5 is widely used in scientific computing for storing **large, complex, hierarchical datasets** efficiently.  
It works like a **file system inside a single file**, with folders (groups) and files (datasets).

#### Why use HDF5 via `h5py`?
- Handles **large datasets** that don’t fit into memory (out-of-core).  
- Supports **hierarchical storage**: groups (like directories) and datasets (like files).  
- Stores both **data** and **metadata** (attributes).  
- Efficient for **scientific data pipelines**, machine learning, and HPC.

#### Common Features (with links)
- **Create a new HDF5 file**: `h5py.File("data.h5", "w")`  
  Docs: <https://docs.h5py.org/en/stable/high/file.html>  
- **Create groups** (like folders): `f.create_group("experiment1")`  
  Docs: <https://docs.h5py.org/en/stable/high/group.html>  
- **Create datasets**: `f.create_dataset("array", data=numpy_array)`  
  Docs: <https://docs.h5py.org/en/stable/high/dataset.html>  
- **Access attributes**: `dset.attrs["units"] = "cm"`  
  Docs: <https://docs.h5py.org/en/stable/high/attr.html>  

#### Example Use Cases
- Storing large NumPy arrays (e.g., satellite imagery, simulation data).  
- Organizing experimental results (multiple groups for trials).  
- Saving deep learning datasets (used by TensorFlow, PyTorch).  
- Sharing scientific data across different programming languages (HDF5 is language-independent).  

> Full docs: <https://docs.h5py.org/en/stable/>


```python
import h5py

# -----------------------------
# 1) Create a new HDF5 file
# -----------------------------
with h5py.File("example.h5", "w") as f:
    # Create a group (like a folder)
    grp = f.create_group("experiment1")

    # Create a dataset (like a file inside the folder)
    data = np.arange(20).reshape(4, 5)
    dset = grp.create_dataset("measurements", data=data)

    # Add an attribute (metadata) to the dataset
    dset.attrs["units"] = "cm"
```

```python
# -----------------------------
# 2) Reopen the file and explore
# -----------------------------
with h5py.File("example.h5", "r") as f:
    print("\nTop-level keys (groups) = ", list(f.keys()))

    # Access group
    grp = f["experiment1"]
    print("\nKeys inside 'experiment1' group = ", list(grp.keys()))

    # Access dataset
    dset = grp["measurements"]
    print("\nDataset shape = ", dset.shape)
    print("\nDataset dtype = ", dset.dtype)

    # Show actual data
    print("\nDataset values = \n", dset[...])

    # Access attributes
    print("\nDataset attributes = ", dict(dset.attrs))
```
